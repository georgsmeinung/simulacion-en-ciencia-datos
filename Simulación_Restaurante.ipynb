```python
# Cargar librerías
import simpy
```

```python
def mesero(env):
  while True:
    print(f"Se toma la orden a los comensales en el segundo {env.now}")
    toma_orden = 5
    yield env.timeout(toma_orden) # Modelamos la duración

    print(f"Los meseros dan la orden a los cocineros en el segundo {env.now}")
    da_orden = 2
    yield env.timeout(da_orden)

    print(f"Los meseros sirven la comida a los comensales en el segundo {env.now} \n")
    sirve_orden = 5
    yield env.timeout(sirve_orden)

env = simpy.Environment() # Creamos el entorno de simulación
env.process(mesero(env)) # Ingresamos a la función al entorno de Simulación
env.run(until = 32) # Cuanto tiempo queremos que dure la simulación
```

```python
import simpy
from random import random, seed
```

```python
def cliente(env, nombre, restaurante,**duracion):
  while True:
    yield env.timeout(random()*10) # hay un nuevo cliente dentro de 10 minutos
    print(f"{nombre} entra al restaurante y espera al mesero en el minuto {round(env.now,2)}")
    with restaurante.request() as req:
       yield req

       print(f"Hay lugares disponibles {nombre} se sienta en el minuto {round(env.now,2)}")
       yield env.timeout(duracion['sentarse'])

       print(f"{nombre} comienza a ver el menú en el minuto {round(env.now,2)}")
       yield env.timeout(duracion['elige_comida'])

       print(f"Los meseros toman la orden del {nombre} en el minuto {round(env.now,2)}")
       yield env.timeout(duracion['da_orden'])

       print(f"{nombre} espera su comida en el minuto {round(env.now,2)}")
       yield env.timeout(duracion['espera_comida'])

       print(f"{nombre} empieza a comer en el minuto {round(env.now,2)}")
       yield env.timeout(duracion['come'])

       print(f"{nombre} empieza a pagar en el minuto {round(env.now,2)}")
       yield env.timeout(duracion['paga'])

       print(f"{nombre} se va en el minuto {round(env.now,2)}")
```

```python
seed(1)
env = simpy.Environment()
restaurante = simpy.Resource(env, capacity = 10)
tiempos = {"sentarse": 1, "elige_comida": 10, "da_orden":5, "espera_comida": 20, "come": 45, "paga": 10}

# Simular 12 clientes
n = 12
for i in range(n):
  env.process(cliente(env,f'Cliente {i}', restaurante, **tiempos))
env.run(until = 200)
```
